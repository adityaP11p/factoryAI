```python
# loading cleaned data 
df = pd.read_csv('../data/processed/cleaned.csv')
df.head()
```

```python
df = df.sort_values(by='timestamp')
```

```python
# rolling mean 
df['air_temp_mean_6'] = df['Air temperature [K]'].rolling(6).mean()
```

```python
df['process_temp_mean_6'] = df['Process temperature [K]'].rolling(6).mean()
```

```python
# rolling standard deviation 
df['rpm_std_6'] = df['Rotational speed [rpm]'].rolling(6).std()
```

```python
df['torque_std_6'] = df['Torque [Nm]'].rolling(6).std()
```

```python
df['air_temp_lag1'] = df['Air temperature [K]'].shift(1)
```

```python
df['air_temp_lag2'] = df['Air temperature [K]'].shift(2)
```

```python
df['temp_diff'] = df['Process temperature [K]'] - df['Air temperature [K]']
```

```python
df['wear_torque'] = df['Tool wear [min]'] * df['Torque [Nm]']
```

```python
df = df.dropna()
```

```python
# saving
df.to_csv('../data/features/features.csv',index=False)
```

```python
df = pd.read_csv("../data/features/features.csv")
```
